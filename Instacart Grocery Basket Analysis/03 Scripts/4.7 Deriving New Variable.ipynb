# 4) Creating the 'price_label' and 'busiest_day' columns

```python
# Working with a subset of the dataframe for now—the first one million rows
df = df_ords_prods_mergd[:1000000]
```

```python
# Creating a flag that sorts products in my ords_prods_merged dataframe according to price - a user-defined funtion (price_label)

def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
# Apply the function
df['price_range'] = df.apply(price_label, axis=1)
```

```python
# Frequency of a new column price_label - checking the values in a new column
df['price_range'].value_counts(dropna = False)
```

```python
#checking what the most expensive item is in the subset
df['prices'].max()
```

```python
# Using If-Statements with the loc() Function for accomplishing the same thing above
df.loc[df['prices'] > 15, 'price_range_loc'] = 'High-range product'
```

```python
df.loc[(df['prices'] <= 15) & (df['prices'] > 5), 'price_range_loc'] = 'Mid-range product' 
```

```python
df.loc[df['prices'] <= 5, 'price_range_loc'] = 'Low-range product'
```

```python
df['price_range_loc'].value_counts(dropna = False)
```

# 5) Suppose your clients have changed their minds about the labels you created in your “busiest_day” column. Now, they want “Busiest day” to become “Busiest days” (plural). This label should correspond with the two busiest days of the week as opposed to the single busiest day. At the same time, they’d also like to know the two slowest days. Create a new column for this using a suitable method.

```python
# Creating a new column that summarizes how busy each day of the week is
df_ords_prods_mergd['order_dow'].value_counts(dropna = False)
```

```python
result1 = []

for value in df_ords_prods_mergd["order_dow"]:
  if value == 0 and 1:
    result1.append("Busiest days")
  elif value == 3 and 4:
    result1.append("Slowest days")
  else:
    result1.append("Regularly busy")
```

```python
# creating new column busiest days
df_ords_prods_mergd['busiest_days'] = result1
```
